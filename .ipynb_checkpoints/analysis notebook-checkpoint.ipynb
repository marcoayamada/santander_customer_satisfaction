## Get metadata from dataset

This dataset has more than 200 features and this will help to slice and separate the data.

````python
def get_meta(train):
    """
    Organizing dataframe in 4 attributes:
    - role
    - level
    - keep
    - dtype

    Will be util when categorical and numerical data used.
    """
    data=[]
    for col in train.columns:
        # role
        if col=='TARGET':
            role='target'
        elif col=='ID':
            role='id'
        else:
            role='input'
            
        # level - identified in behind cell
        if 'ind_var' in col or col == 'TARGET':
            level='binary'
        elif col=='ID':
            level='nominal'
        elif train[col].dtype == np.float64: # float64 are numeric
            level = 'interval'
        elif train[col].dtype == np.int64:  # int64 are categorical
            level = 'ordinal'  
    
        # keep
        keep = True
        if col == 'ID':
            keep = False
        if col =='TARGET':
            keep=False
            
        # data type 
        dtype = train[col].dtype
        
        col_dict={
            'varname': col,
            'role'   :role,
            'level'  :level,
            'keep'   :keep,
            'dtype'  :dtype
        }
        data.append(col_dict)
    meta = pd.DataFrame(data, columns=['varname', 'role', 'level', 'keep', 'dtype'])
    meta.set_index('varname', inplace=True)
    
    return meta
````

### Strong correlations <0.7

````python
# - Breaking dataframe in series
# - Removing auto correlation
# - Filtering with correlations > 0.5 and sorting values by correlation
# - Add column with sorted level_0 and level_1 merged as str
#     - This is necessary to remove level_0: aaa, level_1: bbb; level_0: bbb, level_1: aaa, for example.
#     - Apply drop_duplicates() after apply function don't work correctly. Compare columns in df works great.
# - Applying drop_duplicates() in pandas df and drop aux column aggstr

corr = df[col_interval].corr().stack()
corr = corr[corr.index.get_level_values(0) != corr.index.get_level_values(1)]
corr = corr[corr > 0.7].reset_index().sort_values(by=0, ascending=False)
corr['aggstr'] = corr.apply(lambda x: str(sorted([x.level_0, x.level_1])), axis=1)
corr = corr.drop_duplicates(subset='aggstr').sort_values(by=0, ascending=False).drop('aggstr', axis=1)
````

## Base models

Using all features and basic hyperparameters.

````python
def plot_result_barh(result, width=None, figsize=None):
    """
    Plot list of results
    """    
    ax = pd.DataFrame.from_records(result, columns=['Classifier', 'Mean', 'Std'], index='Classifier')\
                 .sort_values(by='Mean')\
                 .plot(kind='barh', width=width, figsize=figsize)

    # set individual bar lables using above list
    for i in ax.patches:
        # get_width pulls left or right; get_y pushes up or down
        ax.text(i.get_width()+0.005, i.get_y()+0.1, \
                str(round((i.get_width()), 2)), fontsize=10, color='dimgrey')

    plt.xticks(np.arange(0, 1.1, step=0.1))
    ax.legend(loc='lower right')
````

````python
def cross_val_model(x, y, model):
    """
    First, model trains with folds from skf and then cross_val_score() uses same
    model trained (it's cloned) to calculate your score using your cv. Seems
    confused, but works great.

    return: str(model), cross_score.mean(), cross_score.std() between all folds (skf and cross_score)
    """
    x = np.array(x)
    y = np.array(y)
    
    skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=99) # return index of x,y
    
    cross_score_=[] # save all cross_score
    for i, (t1_index, t2_index) in enumerate(skf.split(x, y)):
        xt1, xt2 = x[t1_index], x[t2_index]
        yt1, yt2 = y[t1_index], y[t2_index]
        
        #print ("Fit %s fold %d" % (str(model).split('(')[0], i))
        model.fit(xt1, yt1)
        cross_score = cross_val_score(model, xt2, yt2, cv=3, scoring='roc_auc', n_jobs=-1)
        #print("    cross_score: %.5f" % cross_score.mean())
        cross_score_.append(cross_score.mean())
    
    return (str(str(model).split('(')[0]), np.array(cross_score_).mean(), np.array(cross_score_).std())
````

````python
# List of emsemble params

emsemble_params = {}
emsemble_params['n_estimators'] = 200
emsemble_params['max_depth'] = 6
emsemble_params['min_samples_split'] = 70
emsemble_params['min_samples_leaf'] = 30
emsemble_params['random_state'] = 99
emsemble_params['class_weight']='balanced'
emsemble_params['n_jobs']=-1
````

````python
# List of algortihms to test

algs = [
    LinearSVC(random_state=99, class_weight='balanced'),
    LogisticRegression(random_state=99, class_weight='balanced'),
    ExtraTreesClassifier(**emsemble_params),
    RandomForestClassifier(**emsemble_params)
]
````

````python
x = df[cols]
y = df.TARGET
````

````python
# Add result in a list
result=[]
for clf in algs:
    fnl = cross_val_model(x, y ,clf)
    result.append(fnl)
````

````python
plot_result_barh(result, 0.7)
cfg_plot('Baseline models (all features)', 'ROC AUC score')
````

## Testing by features types

````python
# Comparing ordinal+binary and interval features between all algorithms. The if in second 'for' serves to
# add postfix in the algorithm description.

result=[]
y = df.TARGET
for i, feat in enumerate([col_ordinal+col_binary, col_interval]):
    x = df[feat]
    for clf in algs:
        alg_, mean_, std_ = cross_val_model(x, y ,clf)
        
        if i==0:
            alg_ = alg_+' [ord+bin]'
        else:
            alg_ = alg_+' [interval]'
    
        result.append((alg_, mean_, std_))
````

````python
plot_result_barh(result, 0.7, (15,8))
cfg_plot('Testing ordinal+binary and interval features', 'ROC AUC score')
````

## Testing with 30 most correlationed features

Testing 30 features most correlationed with 4 algorithms.

````python
best_30 = get_feature_importances(cols, rf.feature_importances_, 30).index
````

````python
"""
Testing all algorithms and the interval of 5 and 30 best features.
"""
result=[]
for clf in algs:
    for n_feat in [5, 10, 15, 20, 25, 30]:  
        x = df[best_30[:n_feat]]
        y = df.TARGET

        alg_, mean_, std_ = cross_val_model(x, y ,clf)
        
        alg_ = alg_ + ' [{0}]'.format(n_feat)
        result.append((alg_, mean_, std_))
````

````python
plot_result_barh(result, 0.7, (13,13))
cfg_plot('30 best features comparation', 'ROC AUC score')
````

## Testing features with One Hot Encoding

````python
def one_hot_encoder(df, one_hot, limit):
    df_ = df.copy()
    for c in one_hot:
        if len(one_hot[c])>2 and len(one_hot[c]) < limit:
            for val in one_hot[c]:
                df_[c+'_oh_' + str(val)] = (df_[c].values == val).astype(np.int)
    return df_
````

````python
one_hot = {c: list(df[c].unique()) for c in df.columns if c not in ['TARGET', 'ID']}
````

````python
df_ohe = one_hot_encoder(df, one_hot, 7)
````

````python
print(str(df_ohe.drop(['TARGET', 'ID'], axis=1).columns.size)+' columns')
````

````python
x = df_ohe.drop(['TARGET', 'ID'], axis=1)
y = df_ohe.TARGET

result=[]
for clf in algs:
    fnl = cross_val_model(x, y ,clf)
    result.append(fnl)
````

````python
plot_result_barh(result, 0.7)
cfg_plot('Comparation between one hot encoded features', 'ROC AUC score')
````

No relative difference between original features
